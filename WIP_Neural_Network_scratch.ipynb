```julia
function sigmoid_LA(X)
    return 1 ./(1 .+ exp.(-X))
end
```

```julia
function log_loss(ytrain, y_predicted)
    y_predicted_new = Array(y_predicted)
    ep = 1e-15
    for i in eachindex(y_predicted_new)
        if y_predicted_new[i] == 0
        y_predicted[i] = ep
        elseif y_predicted_new[i] == 1
        y_predicted_new[i] = 1-ep
        end
    end
    println("log ", typeof(y_predicted_new), typeof(ytrain))
    return -mean(ytrain.*log.(y_predicted_new) .+(1 .-ytrain).*log.(1 .- y_predicted_new)) # maybe convert y_pred into an array
end
```

```julia
function grad_dec(x1, x2, x3, x4, x5, x6, x7, x8, ytrain, epochs, loss_thresold)
    w1=w2=w3=w4=w5=w6=w7=w8=w9=2
    bias = 1
    rate = 4
    n = length(x1)
    for i in 1:epochs
        weighted_sum = w1.*x1 + w2.*x2 + w3.*x3 + w4.*x4 + w5.*x5 + w6.*x6 + w7.*x7 + w8.*x8 .+ bias # not an array :(
                                            # println(size(weighted_sum))
        y_predicted= Relu(weighted_sum)
                                            # print(size(y_predicted)) 
        loss = log_loss(ytrain,y_predicted)
        print(loss)
                                            # loss = Flux.binarycrossentropy(y_predicted, ytrain;  ϵ = eps.(y_predicted));
        j = y_predicted - ytrain
    
        dw1 = (1/n).*dot(transpose(x1), j)
        dw2 = (1/n).*dot(transpose(x2), j)
        dw3 = (1/n).*dot(transpose(x3), j)
        dw4 = (1/n).*dot(transpose(x4), j)
        dw5 = (1/n).*dot(transpose(x5), j)
        dw6 = (1/n).*dot(transpose(x6), j)
        dw7 = (1/n).*dot(transpose(x7), j)
        dw8 = (1/n).*dot(transpose(x8), j)
        dbias = mean(j)
        
        w1 = w1 - dw1.*rate
        w2 = w2 - dw2.*rate
        w3 = w3 - dw3.*rate
        w3 = w3 - dw3.*rate
        w4 = w4 - dw4.*rate
        w5 = w5 - dw5.*rate
        w6 = w6 - dw6.*rate
        w7 = w7 - dw7.*rate
        w8 = w8 - dw8.*rate
        bias = bias - dbias*rate
        println("Epoch=",i," w1=",w1, " w2=",w2," w3=",w3," w4=",w4," w5=",w5," w6=",w6," w7=",w7," w8",w8," bias=",bias," loss=",loss) 
        if loss <=loss_thresold
            break
        end
    end
return w1,w2,w3, w4, w5, w6, w7, w8,bias
end
```

```julia
g = []
g = grad_dec(x1, x2, x3, x4, x5, x6, x7, x8, ytrain, 1000, 0.4)
```

```julia
prediction = []
weighted_sum = g[1].*y1 + g[2].*y2 + g[3].*y3 + g[4].*y4 + g[5].*y5 + g[6].*y6 + g[7].*y7 + g[8].*y8 .+ g[9]
prediction =sigmoid_LA(weighted_sum)

for i in eachindex(prediction) 
    if prediction[i] >0.5
        prediction[i] =1
    else
        prediction[i] =0
    end
end
```
